1. loading the dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('creditcard.csv')
df.head()
```

2. data processing

```python
df.shape
df.info()
df.isnull().sum()
df['Class'].value_counts(normalize=True) * 100
```

3. visualization

```python

sns.countplot(x='Class', data=df)
plt.title('Class Distribution (0: Legit | 1: Fraud)')
plt.show()
```

5. preprocessing

```python
print(df.columns.tolist())
from sklearn.preprocessing import StandardScaler

# Standardize 'Amount' and 'Time' (optional but common)
df['scaled_amount'] = StandardScaler().fit_transform(df[['Amount']])
df['scaled_time'] = StandardScaler().fit_transform(df[['Time']])


df.drop(['Amount', 'Time'], axis=1, inplace=True)

scaled_df = df[['scaled_time', 'scaled_amount'] + [f'V{i}' for i in range(1, 29)] + ['Class']]
```

6. train and testing multiple models

```python
from sklearn.model_selection import train_test_split

X = scaled_df.drop('Class', axis=1)
y = scaled_df['Class']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
lr.fit(X_train, y_train)
```

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)
```

```python
from sklearn.neural_network import MLPClassifier

mlp = MLPClassifier(hidden_layer_sizes=(32, 16), max_iter=500)
mlp.fit(X_train, y_train)
```

7. evaluating the models

```python
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score

def evaluate(model, X_test, y_test):
    y_pred = model.predict(X_test)
    print(confusion_matrix(y_test, y_pred))
    print(classification_report(y_test, y_pred))
    print("ROC AUC Score:", roc_auc_score(y_test, y_pred))


print("Logistic Regression:")
evaluate(lr, X_test, y_test)

print("\nDecision Tree:")
evaluate(dt, X_test, y_test)

print("\nNeural Network:")
evaluate(mlp, X_test, y_test)
```

8. visualizing the results.

```python
from sklearn.metrics import RocCurveDisplay

RocCurveDisplay.from_estimator(lr, X_test, y_test)
plt.title("ROC Curve - Logistic Regression")
plt.show()
```
